## Define the `randn_plot` function.

```julia
"""
randn_plot(μ, σ, n_sample, bins)
Plot the histogram of random normal distribution 
with the mean μ and width σ.
n_sample is number of samples to be generated and 
n_bin is the number of bins.
"""
function randn_plot(μ = 0, σ = 1, n_sample = 100_000, n_bin = 100)
    samples = μ .+ σ * randn(n_sample)
    println("The mean value is ", mean(samples), " and the std is ", std(samples), ".")
    histogram(samples, 
        bins = n_bin,
        xlabel = "x",
        ylabel = "counts", 
        label = "μ = $μ, σ = $σ",
        title = "Histogram of random normal distribution."
    )
end

# test this function
# randn_plot()
```

## Case 1: $ \mu = 0, \sigma = 1$

```julia
randn_plot(0, 1)
```

## Case 2: $ \mu = 0, \sigma = 2$

```julia
randn_plot(0, 2)
```

## Case 3: $ \mu = 0, \sigma = \sqrt{2}$

```julia
randn_plot(0, √2)
```

## Case 4: $ \mu = 2, \sigma = \sqrt{2}$

```julia
randn_plot(2, √2)
```
