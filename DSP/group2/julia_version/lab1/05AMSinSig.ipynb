```julia
# load packages
using Plots
using FFTW
import LinearAlgebra:norm
```

## Define the amplitude modulated (AM) sinusoid signal.
The signal is given by
$$
s(t) = A \cos(2\pi f_1 t) \times \sin(f_0 t + \phi_0),
$$
where $A$ is the $snr$.

```julia
"""
AMSinSig(dataT, snr, f0, f1, ϕ0) generates a amplitude modulated (AM) 
sinusoid S. dataT is the vector of time stamps at which the samples 
of the signal are to be computed. snr is the matched filtering 
signal-to-noise ratio of S, and [f0, f1, ϕ0] are three coefficients that 
parametrize the phase of the signal.
"""
function AMSinSig(dataT, snr::Real, f0::Real, f1::Real, ϕ0::Real)
    phaseVec01 = 2π*f1*dataT
    phaseVec02 = f0*dataT .+ ϕ0
    sigVec = cos.(phaseVec01) .* sin.(phaseVec02)
    sigVec = snr*sigVec/norm(sigVec)
end

# test the function
AMSinSig(1, 1, 1, 1, 1), AMSinSig([1,2], 1, 1, 1, 1)
```

```julia
?AMSinSig
```

## Plot the signal.

```julia
#Set signal parameters.
snr = 10;
f0 = 20;
f1 = 20;
ϕ0 = π;
```

```julia
#Instantaneous frequency after 1 sec.
maxFreq = max(2π*f1, f0+ϕ0);
samplFreq = 5*maxFreq;
samplIntrvl = 1/samplFreq;
```

```julia
#Time samples.
times = collect(0:samplIntrvl:1.0);

# Number of samples
nSamples = length(times);

# Generate the signal.
signals = AMSinSig(times, snr, f0, f1, ϕ0);
```

### Plot the signal.

```julia
plot(times,signals)
```

### Plot the periodogram.

```julia
# Length of data 
dataLen = times[end]-times[1]

# DFT sample corresponding to Nyquist frequency
kNyq = Int(floor(nSamples/2)+1)

# Positive Fourier frequencies
posFreq = (0:(kNyq-1))*(1/dataLen);

# FFT of signal
fftSig = fft(signals);

# Discard negative frequencies
fftSig = fftSig[1:kNyq];
```

```julia
# Plot periodogram
plot(posFreq,abs.(fftSig))
```
