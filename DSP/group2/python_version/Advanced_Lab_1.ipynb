# Time-Frequency Analysis(Q-Transformation)

```python
%matplotlib notebook
%config InlineBackend.figure_format = 'retina'

from signalform import SignalForm, test_signal
import matplotlib.pyplot as plt
import numpy as np

sampFreq = 1024
nSamples = 2048
timeVec = np.arange(nSamples)/sampFreq;

SigForm = SignalForm(signal='AM-FM-sine', A=10, b=10, f_0=10, f_1=10)
sigVec = SigForm(timeVec)

test_signal(timeVec, sigVec, title='AM-FM-sine')
```

## Spectrogram

```python
from scipy import signal
f, t, Sxx = signal.spectrogram(x=sigVec, fs=sampFreq, window='hanning', 
                               nperseg=256, noverlap=250)

plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
