```julia
using Statistics
using Plots
pyplot();
```

```julia
n_sample = 100_000;
```

#### $\mu = 0$, $\sigma = 1$

```julia
samples = randn(n_sample);
println("The mean value is ", mean(samples), " and the std is ", std(samples), ".")
histogram(samples, bins=:scott)
```

#### $\mu = 0$, $\sigma = 2$

```julia
samples = 2 * randn(n_sample);
println("The mean value is ", mean(samples), " and the std is ", std(samples), ".")
histogram(samples, bins=:scott)
```
